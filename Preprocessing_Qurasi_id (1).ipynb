```python
def casefolding(text):
    if isinstance(text, str):
        text = text.lower()
    return text

# Assuming df is your DataFrame
df['synopsis'] = df['synopsis'].apply(casefolding)
```

```python
def clean(text):
    text = str(text)
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'@[A-Za-z0-9]+', '', text)  # Remove @mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text)  # Remove hashtags
    text = re.sub(r'rt[\s]', '', text)  # Remove RT
    text = re.sub(r'diterjemahkan oleh google', '', text)  # Remove specific phrase
    text = re.sub(r'user', '', text)  # Remove the word 'user'
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[0-9]+', '', text)  # Remove numbers
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)  # Remove single characters
    text = text.replace('\n', ' ')  # Replace newline with space
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace

    # Remove emojis
    text = re.sub(r'[^\w\s]', '', text)

    # Remove HTML entities
    text = html.unescape(text)

    # Normalize elongated words and repeated sequences
    text = re.sub(r'\b\w*(\w)\1{2,}\w*\b', '', text)

    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)

    return text

df['synopsis']  = df['synopsis'].apply(clean)
```

```python
def stopsastra(text):
  factory = StopWordRemoverFactory()
  stopword = factory.create_stop_word_remover()
  text = stopword.remove(text)
  return text

df['synopsis'] = df['synopsis'].apply(stopsastra)
```

```python
def konversi_slangword(kalimat, kamus):
  kalimatbaru=''
  if len(kalimat)>0:
    for kata in kalimat:
      katabaru= kata.strip()
      for i in range(len(kamus)):
        if katabaru==kamus[i][0]:
          katabaru=kamus[i][1]
      kalimatbaru=kalimatbaru+' '+katabaru
  return kalimatbaru

df['synopsis'] = df['synopsis'].apply(konversi_slangword, kamus=kamus)
```
